```python
import pymn
import pandas as pd
import matplotlib.pyplot as plt
import scanpy as sc
import anndata as ad
import numpy as np
import itertools
import scipy.stats
import seaborn as sns
```

```python
norm_data = sc.read_h5ad('/data/passala/Collaborator_Data/Bruno_Ken_Collab/Post_review_Dataset/faster_reading_full_data.h5ad')
```

```python
#maize_markers = pd.read_csv('/data/passala/Collaborator_Data/Bruno_Ken_Collab/Post_review_Dataset/maize_high_purity_allMarkers.csv', index_col= 0)
#maize_markers = pd.read_csv('/data/passala/Collaborator_Data/Bruno_Ken_Collab/Post_review_Dataset/setaria_high_purity_markers.csv', index_col= 0)
maize_markers = pd.read_csv('/data/passala/Collaborator_Data/Bruno_Ken_Collab/Post_review_Dataset/sorghum_high_purity_markers.csv', index_col= 0)

maize_markers = maize_markers.set_index('gene').drop(columns= 'X')
maize_markers
```

```python
panda_of_expression_results = pd.DataFrame(index = norm_data.obs.index, columns= norm_data.var.index, data = norm_data.X )
panda_of_expression_results
```

```python
panda_of_expression_results = panda_of_expression_results.T
panda_of_expression_results
```

```python
panda_of_expression_results = panda_of_expression_results.merge(right = maize_markers, left_index= True, right_index= True)
```

```python
expression_results_by_gene = panda_of_expression_results.drop(columns = ['p_val','avg_log2FC','pct.1','pct.2','p_val_adj','cluster'])
expression_results_by_gene
```

```python
spec_weighted_markers = expression_results_by_gene.multiply(panda_of_expression_results['avg_log2FC'].values, axis = 0)
spec_weighted_markers
```

```python
labeled_spec_weighted_markers = spec_weighted_markers.copy()
labeled_spec_weighted_markers['cluster'] = panda_of_expression_results['cluster']
labeled_spec_weighted_markers
```

```python
calculated_average_pre_weight_ici =labeled_spec_weighted_markers.groupby('cluster').mean()
calculated_average_pre_weight_ici
```

```python
binary_mask_of_spec_weighted = spec_weighted_markers>0
binary_mask_of_spec_weighted['cluster'] = panda_of_expression_results['cluster']
proportion_of_markers_present = binary_mask_of_spec_weighted.groupby('cluster').mean()
```

```python
proportion_of_markers_present
```

```python
full_ici_scores_unnormalized = calculated_average_pre_weight_ici*proportion_of_markers_present
full_ici_scores_unnormalized
```

```python
full_ici_scores = full_ici_scores_unnormalized/full_ici_scores_unnormalized.sum() 
full_ici_scores
```

```python
full_ici_scores = full_ici_scores.T
```

```python
labeled = 'Sorghum_' + full_ici_scores.columns.astype(str)
```

```python
full_ici_scores.columns = labeled
full_ici_scores
```

```python
full_ici_scores.to_csv('/data/passala/Collaborator_Data/Bruno_Ken_Collab/Post_review_Dataset/sorghum_based_ici_scores.csv')
```
